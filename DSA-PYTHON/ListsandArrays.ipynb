```python
# List Methods
# Python lists come with several built-in algorithms (called methods), to perform common operations like appending, sorting, and more.
```

```python
new_list = [91,82,73,64, 45]
print(new_list)
print(type(new_list))

updated_list = [1, 2.5, "Hello", True, [1, 2, 3]]
print(updated_list)
print(type(updated_list))

new_list.append(999)
print(new_list)

new_list.sort()
print(new_list)
```

```python
my_list = [1, 2, 3]

# Add elements
my_list.append(4)           # [1, 2, 3, 4] - add to end
my_list.insert(1, 1.5)      # [1, 1.5, 2, 3] - insert at index
my_list.extend([5, 6])      # [1, 2, 3, 5, 6] - add multiple elements
```

```python
# Remove elements
my_list.remove(2)           # [1, 3] - remove first occurrence
popped = my_list.pop()      # removes and returns last element
popped = my_list.pop(1)     # removes and returns element at index
my_list.clear()             # [] - remove all elements
```

```python
# Searching/Information
my_list =[1, 2, 2, 2, 4]

count = my_list.count(2)      # 2 - count occurrences
print(count)

index = my_list.index(4)      # 3 - find index of first occurrence
print(index)

length = len(my_list)
print(length)

exists = 3 in my_list  # True - check existence
print(exists)
```

```python
# Searching/Information
new_list = [9,8,7,6,5,4,3,2,1]
print(new_list)

new_list.sort() # Sorts the list in ascending order
print(new_list)


new_list.sort(reverse=False) # Reverses the list
print(new_list)


new_list.sort(reverse=True) # Reverses the list
print(new_list)

new_list.reverse() # Reverses the list
print(new_list)

new_list.sort(reverse=True) # Reverses the list
print(new_list)

sorted_list = sorted(new_list)  # returns new sorted list
print(sorted_list)
```

```python
# Array Methods (array module)

from array import array
arr = array('i',[1,2,3,4])
print(arr)

# Most list methods work on arrays too
arr.append(99)
arr.append(98)
arr.append(97)
arr.append(96)
arr.extend([6, 7])

print(arr)

print(arr.pop())
print(arr.pop())

print(arr.remove(1))
arr.fromlist([8, 9])# append from list
print(arr)

lst = arr.tolist()          # convert to list
print(lst)

arr_bytes = arr.tobytes()   # convert to bytes
print(arr_bytes)
```

```python
# Core List Methods
append()
pop()
len()
sort()
index()
insert()
remove()
clear()
copy()
count()
reverse()

# Queue/Deque Methods
popleft()           # from collections import deque
appendleft()

# Heap/Priority Queue
heappush()          # import heapq
heappop()
heapify()

# String Methods (for string problems)
split()
join()
strip()
find()
replace()
lower()
upper()

# Set Methods
add()
remove()
discard()
union()    # | operator
intersection()    # & operator
difference()    # - operator

# Dictionary Methods
keys()
values()
items()
get()
pop()
setdefault()

# Critical Operations (not methods but essential)
# Slicing
arr[start:end:step]

# Membership
x in arr

# Comprehension
[x for x in arr if condition]
```

```python
# Create an algorithm to find the lowest value in a list

my_array = [7,1,3,99,5,9,2]

minVal = my_array[0]

for i in my_array:
    if i < minVal:
        minVal = i

print("Minimum value is:", minVal)
```
